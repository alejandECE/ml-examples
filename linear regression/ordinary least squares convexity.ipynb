```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as alg
```

```python
dataset = pd.read_csv('../../datasets/others/tremors/tremors.csv')
dataset = dataset.dropna()
dataset.head()
```

```python
pts = 25
w0 = np.linspace(-5,5,pts)
w1 = np.linspace(-5,5,pts)
w0,w1 = np.meshgrid(w0,w1)
```

```python
W = np.hstack((w0.reshape(pts*pts,1),w1.reshape(pts*pts,1)))
x = dataset['amplitude'].values.reshape(len(dataset),1)
y = dataset['fuzzy score'].values.reshape(len(dataset),1)
X = np.hstack((np.ones((len(x),1)),x))
J = (0.5*((X.dot(W.T) - y)**2).sum(axis=0)).reshape(pts,pts)
```

```python
w = alg.pinv(X).dot(y) # mle
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.scatter(x,y,c='b')
plt.plot(np.array([x.min(),x.max()]),w[1]*np.array([x.min(),x.max()]) + w[0],'k')
plt.title('Linear Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.subplot(122)
plt.contour(w0,w1,J,levels=25)
plt.plot(w[0],w[1],'rx')
plt.title('Cost function')
plt.xlabel('intercept (w0)')
plt.ylabel('slope (w1)')
plt.grid()
```
