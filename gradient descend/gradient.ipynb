```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from numerical_gradient import approximate_gradient
```

```python
# Building mesh
pts = 50
x = np.linspace(-4,4,pts)
y = np.linspace(-4,4,pts)
[x,y] = np.meshgrid(x,y)
spacing = (x.max() - x.min())/pts
```

```python
def graph_gradient(x,y,z,gstep=3):
    fig = plt.figure(figsize=(15,6))
    ax = fig.add_subplot(121, projection='3d')
    dx,dy = approximate_gradient(z, spacing)
    ax.plot_surface(x,y,z,cmap='viridis')
    ax = fig.add_subplot(122)
    ax.contour(x,y,z,levels=12)
    ax.quiver(x[::gstep,::gstep],y[::gstep,::gstep],-dx[::gstep,::gstep],-dy[::gstep,::gstep])
    ax.grid()
```

```python
# Function 1
z = x**2+y**2
graph_gradient(x,y,z)
```

```python
# Function 2
z = x*np.exp(-x**2-y**2)
graph_gradient(x,y,z,gstep=1)
```

```python
# Function 3
z = y**2 - x**2
graph_gradient(x,y,z)
```

```python
# Function 4
z = np.sin(x) + np.cos(y)
graph_gradient(x,y,z)
```

```python
# Function 5
z = y*np.sin(x) + x*np.cos(y)
graph_gradient(x,y,z,gstep=2)
```
